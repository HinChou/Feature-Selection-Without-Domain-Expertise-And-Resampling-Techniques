```python
import numpy as np
import pandas as pd
from matplotlib import style, pyplot as plt
style.use('ggplot')
%matplotlib inline
```

```python
# Load data
cells_infected = pd.read_csv('C:/Users/hinnc/Desktop/cells_infected.csv', header = 0)
```

```python
# Check missing data
print('The Number of Missing Data:', cells_infected.isnull().sum().sum())
```

```python
# Drop the useless column based on the data dictionary
cells_infected.drop(['row', 'col'], axis= 1, inplace = True)

# Make a label based on Intensity_MedianIntensity_OrigGreen_nucleus value, 1 means Infected, 0 means non-infected
cells_infected['Infected_Cell'] = 1
cells_infected.ix[ cells_infected.Intensity_MedianIntensity_OrigGreen_nucleus < 0.03, 'Infected_Cell'] = 0

n_rows, n_cols = cells_infected.shape
print('Dimensionality of Data:')
print('The Number of Rows:', n_rows)
print('The Number of Columns:', n_cols)
```

```python
# Show data
cells_infected.head()
```

```python
# Get all the column names
column_names = cells_infected.columns.tolist()

# Check the number of catogorical columns and numerical columns
cat_cols = []

for column in column_names:
    if (cells_infected[column].dtype != np.float64) & (cells_infected[column].dtype != np.int64):
        cat_cols.append(column)

print('The Number of Catogorical Columns:', len(cat_cols))
print('The Number of Numerical Columns:', len(column_names) - len(cat_cols))

# Seperate feature space and label
X = np.array(cells_infected.drop('Infected_Cell', axis = 1))
y = np.array(cells_infected['Infected_Cell'])
```

## 2. Feature Selection Methods
###  2.1 Removing features with low variance
It is a simple baseline approach to feature selection. It removes all features whose variance doesn’t meet some threshold. By default, it removes all zero-variance features, i.e. features that have the same value in all samples.

```python
from sklearn.feature_selection import VarianceThreshold

# Use default threshold
sel_def = VarianceThreshold()
result = sel_def.fit_transform(X)
print('The Number of Features Selected When Removes All Zero-variance Features:', result.shape[1])

# Use threshold equals 0.2
sel = VarianceThreshold(threshold = 0.2)
result = sel.fit_transform(X)
print('The Number of Features Selected with the Variance Higher Than 0.2:', result.shape[1])
```

```python
# Get the 0 variance featuers
bool_arr = (sel_def.variances_ == 0).tolist() 
seq = [i for i, value in enumerate(bool_arr) if value]
print('0 Variance Featuers:')
for ele in seq:
    print(column_names[ele])
```

Based on the result above, there are 4 features that have the same value in all samples. These features definitely should be removed before building models.

###  2.2 Univariate feature selection
Univariate feature selection works by selecting the best features based on univariate statistical tests. It can be seen as a preprocessing step to an estimator.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# Removing features with 0 variance
column_names.remove('Children_Cells_Count_nucleus')
column_names.remove('Children_Cytoplasm_Count_nucleus')
column_names.remove('Children_FilteredNuclei_Count_nucleus')
column_names.remove('Children_Cytoplasm_Count_cell')

# Seperate feature space and label
X = np.array(cells_infected.drop(['Children_Cells_Count_nucleus', 'Children_Cytoplasm_Count_nucleus', 
                                 'Children_FilteredNuclei_Count_nucleus', 'Children_Cytoplasm_Count_cell',
                                 'Infected_Cell'], axis = 1))
y = np.array(cells_infected['Infected_Cell'])

# Perform a ANOVA test to the samples to retrieve only the three best features as follows
sel = SelectKBest(f_classif, k = 3)
X_new = sel.fit_transform(X, y)
print('The Number of Features Selected:', X_new.shape[1])
```

```python
# Display the names of these three features
new_list = sel.scores_.tolist()

print('Three Best Features Based on ANOVA Test:')

for i in range(3):
    max_current = max(new_list)
    bool_arr = (sel.scores_ == max_current).tolist() 
    for i, value in enumerate(bool_arr):
        if value:
            print(column_names[i])
    new_list.remove(max_current)
```

###  2.3 Recursive Feature Elimination
#### 2.31 Recursive Feature Elimination
Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and weights are assigned to each one of them. Then, features whose absolute weights are the smallest are pruned from the current set features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

# Create the RFE object and retrieve only the three best features as follows
svc = SVC(kernel = 'linear', C = 1)
rfe = RFE(estimator = svc, n_features_to_select = 3, step = 1)
X_new = rfe.fit_transform(X, y)
print('The Number of Features Selected:', X_new.shape[1])
```

```python
# Display the names of these three features
seq = [i for i, value in enumerate(rfe.support_) if value]

print('Three Best Features Based on Recursive Feature Elimination:')

for ele in seq:
    print(column_names[ele])
```

#### 2.32 Recursive Feature Elimination with Cross-validation
Feature ranking with recursive feature elimination and cross-validated selection of the best number of features.

```python
from sklearn.feature_selection import RFECV
from sklearn.svm import SVC

# Create the RFECV object and retrieve the optimal number of features as follows
svc = SVC(kernel = 'linear', C = 1)
rfecv = RFECV(estimator = svc, step = 1, cv = 3, scoring='accuracy')
rfecv.fit(X, y)

print("Optimal Number of Features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure(figsize = (10, 8))
plt.xlabel("Number of Features Selected")
plt.ylabel("Cross Validation Score (Percentage of Correct Classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
```

### Important to Point Out:
The recursive feature elimination with or without cross-validation costs a lot of time the calculate the desired feature space. This is obviously, because as we mentioned before, this procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached. It should be used carefully on the large original feature space, the computational cost is quite expensive.

###  2.4 Feature selection using SelectFromModel
#### 2.41 Tree-based Feature Selection

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

# Create the SelectFromModel object and retrieve the optimal number of features which
# the threshold value is the median of the feature importances.
clf = ExtraTreesClassifier(n_estimators  = 100, n_jobs = -1, random_state = 123)
clf = clf.fit(X, y)
clf.feature_importances_  
model = SelectFromModel(clf, threshold = 'median', prefit=True)
X_new = model.transform(X)
print('The Number of Features Selected:', X_new.shape[1])
```

```python
# Get the feature importances
importances = clf.feature_importances_

std = np.std([tree.feature_importances_ for tree in clf.estimators_],
             axis=0)

# Return the indices of the top 10 important features 
indices = np.argsort(importances)[::-1][:10]
```

```python
# Get the top 10 name of features
features = [column_names[ele] for ele in indices]
```

```python
# Show importance of each variable 
fig = plt.figure(figsize = (10, 8))
axes = plt.subplot2grid((1,1), (0,0))
axes.bar(range(10), importances[indices],
       color = "r", yerr = std[indices], align = "center")
plt.xticks(range(10), features)
plt.xlim((-1, 10))

# Rotate the angle of the labels
for label in axes.xaxis.get_ticklabels():
    label.set_rotation(90)
    
plt.title("Top 10 Important Feature")
plt.xlabel('Name of Variable')
plt.ylabel('Importance')   
```

```python
# Output the top 10 important features 
print("Top 10 Feature Ranking:")
for f in range(10):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

## 3. Comparison and Analysis 
* For the first two feature selection method, it is really self-explanatory

        a. Removing Features with Low Variance: we obtain the desired feature space by delete the low variance features which cannot provide useful information to the models, expecially, when there are 0 variance in the features (features that have the same value in all samples).
        
        b. Univariate Feature Selection: we obtain the desired feature space based on univariate statistical tests.

    Advantages:
        * Easy to understand and interpret
        
        * Fast to calculate

* For Recursive Feature Elimination with Cross-validation method, it performs RFE in a cross-validation loop to find the optimal number of features.
    
    Advantages: 
        * Can actually select the best number of features using cross-validation, does not need to predefine the desired the threshold for the feature space, which might create the bias.
    


* For Tree-based Feature Selection method, we obtain the desired feature space by discarding the irrelevant features.
    
    Advantages: 
        * Probably will have the best trade off between the accuracy of the feature selection and the computational time among all the methods above.
    

## 4. Resampling Techniques
* Random Over-sampling
* Synthetic Minority Over-sampling Technique(SMOTE)

Reference: https://github.com/scikit-learn-contrib/imbalanced-learn/blob/master/examples/plot_unbalanced_dataset.ipynb

```python
from sklearn.decomposition import PCA

from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE

from imblearn.combine import SMOTETomek
from imblearn.combine import SMOTEENN
import seaborn as sns
```

```python
# For visualisation purpose only choose 1000 random sample from orignal data
from numpy.random import choice

# Set a random seed and split the data to training data and testing data
np.random.seed(123)
random_num = choice(np.arange(len(cells_infected)), 1000, replace = False)
random_sample = cells_infected.ix[random_num]

# Seperate feature space and label
X = np.array(random_sample.drop(['Children_Cells_Count_nucleus', 'Children_Cytoplasm_Count_nucleus', 
                                 'Children_FilteredNuclei_Count_nucleus', 'Children_Cytoplasm_Count_cell',
                                 'Infected_Cell'], axis = 1))
y = np.array(random_sample['Infected_Cell'])

print('The Number of Class 1:', sum(random_sample.Infected_Cell == 1))
print('The Number of Class 0:', sum(random_sample.Infected_Cell == 0))
```

```python
# Instanciate a PCA object for the sake of easy visualisation
pca = PCA(n_components=2)

# Fit and transform x to visualise inside a 2D feature space
x_vis = pca.fit_transform(X)

# Plot the original data
# Plot the two classes
palette = sns.color_palette()
plt.scatter(x_vis[y == 0, 0], x_vis[y == 0, 1], label="Class #0", alpha=0.5, 
            edgecolor='k', facecolor=palette[0], linewidth=0.15)
plt.scatter(x_vis[y == 1, 0], x_vis[y == 1, 1], label="Class #1", alpha=0.5, 
            edgecolor='k', facecolor=palette[2], linewidth=0.15)

plt.legend(loc = 3)
```

### Random Over-sampling

```python
# Generate the new dataset using Random Over-sampling method
#verbose = False
ratio = 'auto'
```

```python
# 'Random over-sampling'
OS = RandomOverSampler(ratio = ratio, random_state = 123)
osx, osy = OS.fit_sample(X, y)
```

```python
# 'SMOTE'
smote = SMOTE(ratio = ratio, random_state = 123, kind='regular', n_jobs = -1)
smox, smoy = smote.fit_sample(X, y)
```

```python
# 'SMOTE bordeline 1'
bsmote1 = SMOTE(ratio = ratio, random_state = 123, kind = 'borderline1', n_jobs = -1)
bs1x, bs1y = bsmote1.fit_sample(X, y)
```

```python
# 'SMOTE bordeline 2'
bsmote2 = SMOTE(ratio  = ratio, random_state = 123, kind = 'borderline2', n_jobs = -1)
bs2x, bs2y = bsmote2.fit_sample(X, y)
```

```python
# 'SMOTE SVM'
svm_args={'class_weight': 'balanced'}
svmsmote = SMOTE(ratio = ratio, random_state = 123, kind='svm', **svm_args, n_jobs = -1)
svsx, svsy = svmsmote.fit_sample(X, y)
```

```python
# 'SMOTE Tomek links'
STK = SMOTETomek(ratio = ratio, random_state = 123)
stkx, stky = STK.fit_sample(X, y)
```

```python
# 'SMOTE ENN'
SENN = SMOTEENN(ratio = ratio, random_state = 123)
ennx, enny = SENN.fit_sample(X, y)
```

```python
# Apply PCA to be able to visualise the results
osx_vis = pca.transform(osx)
smox_vis = pca.transform(smox)
bs1x_vis = pca.transform(bs1x)
bs2x_vis = pca.transform(bs2x)
svsx_vis = pca.transform(svsx)
stkx_vis = pca.transform(stkx)
ennx_vis = pca.transform(ennx)
```
